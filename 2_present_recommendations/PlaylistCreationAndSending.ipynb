```python
import pandas as pd
```

```python
df = pd.read_csv('../participant data/survey data/msi_response.csv')
```

```python
table = df.groupby(['user_id', 'item_id', 'value']).sum().reset_index()

usr_mail = table[(table['item_id'] == 'email')]
usr_grp = table[table['item_id'] == 'group']
usr_grp
```

```python
all_users = usr_grp.merge(usr_mail, how='inner', on='user_id').rename(columns={"value_x": "group", "value_y": "email"})[["user_id", "group", "email"]]
# usrs = usr_grp[usr_grp['value'] == 'DUMDUM'].merge(usr_mail, how='inner', on='user_id')
all_users
```

```python
# Get all DUMDUM users, plus a few extra: index 7, 10, 14
user_indexes = (all_users["group"] == "DUMDUM")
user_indexes[14] = True
user_df = all_users[user_indexes]
user_df = user_df[["user_id", "email"]].reset_index(drop=True)
user_df
```

```python
# Import API
import sys
sys.path.append("../modules/")
from API import API
api = API()
```

```python
# Generate playlist names
users = list(user_df['user_id'])
api.generatePlaylistNames(users)
```

```python
# Create playlists for each of our users
api.createAllPlaylistsForAllUsers(users)
```

```python
# Check user playlists
playlists = api.getPlaylists(limit=50)
playlist_names = [playlist['name'] for playlist in playlists]
print(f'users: {len(users)} (should be half the length of playlists)')
print(f'playlists: {len(playlist_names)}') # noice
```

```python
# Create a dictionary where the playlist name is the key and the ID is the value
playlist_ids = [playlist['id'] for playlist in playlists]
playlist_ids

playlist_dict = dict()
for i, name in enumerate(playlist_names):
    playlist_dict[name] = playlist_ids[i]

playlist_dict
```

```python
# Double check 
print(playlist_ids[-1:])
print(playlist_names[-1:]) # this checks out
```

```python
# Get song recommendations for each of our users
diverse = pd.read_csv("../participant data/max_diversified_recommendations_long.csv")
non_diverse = pd.read_csv("../participant data/min_diversified_recommendations_short.csv")

diverse.head()
```

```python
# Populate playlists for our users
for user in users:
    # Get playlist IDs
    diverse_name = user + "_d"
    non_diverse_name = user + "_n"
    diverse_id = playlist_dict[diverse_name]
    non_diverse_id = playlist_dict[non_diverse_name]
    
    # Get songs (pre-sorted by rank_new)
    diverse_songs = list(diverse[diverse['user'] == user]['uri'])
    non_diverse_songs = list(non_diverse[diverse['user'] == user]['uri'])
    
    # Populate the playlists
    api.populatePlaylist(playlist_id= diverse_id, song_uris= diverse_songs)
    api.populatePlaylist(playlist_id= non_diverse_id, song_uris= non_diverse_songs)
```

```python
# Get email addresses of the successfully made playlists
PLAYLISTS = api.getPlaylists(limit=50)
successes = [success['name'] for success in PLAYLISTS]
successes = [success[:-2] for success in successes] # remove the _d/_d at the end
successes = list(set(successes))
print(f"{len(successes)} users had playlists created")
```

```python
sorted(successes)
```

```python
emails = user_df[user_df["user_id"].isin(successes)]
emails
```

```python
# Get playlist URLs for each email
print(PLAYLISTS[0]['name'])
print(PLAYLISTS[0]['external_urls']['spotify'])
name_url = dict()

for item in PLAYLISTS:
    name_url[item['name']] = item['external_urls']['spotify']

name_url
```

```python
def getDiverse(user_id):
    return name_url[user_id + '_d']

def getNonDiverse(user_id):
    return name_url[user_id + '_n']


emails['diverse'] = emails.apply(lambda row: getDiverse(row['user_id']), axis=1)
emails['non_diverse'] = emails.apply(lambda row: getNonDiverse(row['user_id']), axis=1)
emails
```

```python
for i, row in emails.iterrows():
    print(row['email'])
    print(row['diverse'])
    print(row['non_diverse'])
    print()
```
