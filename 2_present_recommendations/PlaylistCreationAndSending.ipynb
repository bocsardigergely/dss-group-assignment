```python
import pandas as pd
```

```python
df = pd.read_csv('../participant data/survey data/msi_response.csv')
```

```python
table = df.groupby(['user_id', 'item_id', 'value']).sum().reset_index()

usr_mail = table[(table['item_id'] == 'email')]
usr_grp = table[table['item_id'] == 'group']
usr_grp
```

```python
all_users = usr_grp.merge(usr_mail, how='inner', on='user_id').rename(columns={"value_x": "group", "value_y": "email"})[["user_id", "group", "email"]]
# usrs = usr_grp[usr_grp['value'] == 'DUMDUM'].merge(usr_mail, how='inner', on='user_id')
all_users
```

```python
# Get all DUMDUM users, plus a few extra: index 7, 10, 14
user_indexes = (all_users["group"] == "DUMDUM")
user_indexes[14] = True
users = all_users[user_indexes]
users = users[["user_id", "email"]].reset_index(drop=True)
users
```
